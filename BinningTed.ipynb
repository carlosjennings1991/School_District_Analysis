```python
# Import Dependencies
import pandas as pd
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
csv_path = "Resources/ted_talks.csv"
ted_df = pd.read_csv(csv_path)

ted_df.head()
```

```python
# Figure out the min views
mins_views = ted_df['comments'].min()
mins_views
```

```python
# find max views

max_views = ted_df['comments'].max()

max_views
```

```python
# Create bins in which to place values based upon TED Talk views
bins = [0, 1000, 2000, 3000, 4000, 5000, 6000, 7000]


# Create labels for these bins
bin_labels = ['quiet', 'not so quiet', 'getting there', 'medium', 'loud', 'louder', 'deafening']
```

```python
# Slice the data and place it into bins
sliced_df = pd.cut(ted_df['comments'], bins, labels=bin_labels)
```

```python
# Place the data series into a new column inside of the DataFrame
ted_df['Comment_Group'] = sliced_df
ted_df
```

```python
# Create a GroupBy object based upon "View Group"
grouped = ted_df.groupby('Comment_Group')

# Find how many rows fall into each bin
grouped["comments"].count()

# Get the average of each column within the GroupBy object
grouped.mean()
```
